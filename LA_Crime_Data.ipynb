## Probability of crime type by premise description

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Crime_Data_from_2020_to_Present.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
contingency_table = pd.crosstab(df['Premis Desc'], df['Crm Cd Desc'], margins=False)

probability_table = contingency_table / contingency_table.sum().sum()
probability_table
```

```python
# top 10 most frequent crime types and premises
top_crimes = df['Crm Cd Desc'].value_counts().head(10).index
top_premises = df['Premis Desc'].value_counts().head(10).index
filtered_df = df[df['Crm Cd Desc'].isin(top_crimes) & df['Premis Desc'].isin(top_premises)]

contingency_table = pd.crosstab(filtered_df['Premis Desc'], filtered_df['Crm Cd Desc'], margins=False)
probability_table = contingency_table / contingency_table.sum().sum()


plt.figure(figsize=(12, 8))
sns.heatmap(probability_table, cmap="YlGnBu", annot=True, fmt=".2f", cbar=True) # found heatmap is part of seaborn library
plt.title('Heatmap of Crime Type Probabilities by Premises Description (Top 10)')
plt.xlabel('Crime Type (Crm Cd Desc)')
plt.ylabel('Premises Description (Premis Desc)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

## Frequency of crime type by area

```python
frequency_table = df.groupby(['AREA NAME', 'Crm Cd Desc']).size().reset_index(name='frequency')
frequency_table
```

```python
# bar plot for frequency of crime types by area
plt.figure(figsize=(12, 8))
sns.barplot(data=frequency_table, x='AREA NAME', y='frequency', hue='Crm Cd Desc')
plt.title('Frequency of Crime Types by Area')
plt.xlabel('Area Name')
plt.ylabel('Frequency of Crimes')
plt.xticks(rotation=45)
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
